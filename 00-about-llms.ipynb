### 00. Om LLM'er (large language model)

De flesta modeller finns både med ett chattliknande gränssnitt och som ett API.

**Några populära komersiella tjänster**

- ChatGPT - Open AI
- Google Gemini - Google (kommer vi att använda idag)
- Claude - Anthropic

**Det finns även gratis modeller som går att köra lokalt**

- llama - Meta
- deepseek - deepseek
- Gemma3 - Google

**Varje tjänst erbjuder flera olika modeller**

Modeller finns i flera generationer och kan vara designade för olika syften och har olika kostnader för användade.
Ju nyare modell dessto dyrare, ofta klarar man sig med äldre, enklare modeller.


| ChatGPT | Google Gemini | Anthropic |  
|---|---|---|
| GPT-4.1 | gemini-2.5-flash-preview-05-20 | Claude Opus 4 |
| o4-mini | gemini-2.0-flash | Claude Sonnet 3.7 |
| o3 | gemini-2.0-flash-lite | Claude Haiku 3.5 |
| GPT-3.5 Turbo | gemini-1.5-flash | ... |
| ... |... |... |

**Kostnaden för användandet av en tjänst beräknas i 'tokens'**
 - en token kan vara ett ord, en del av ett ord eller ett tecken
 - Exempel: Frasen `"ChatGPT är bra!"` blir ungefär 5 tokens: `"Chat", "G", "PT", " är", " bra!"`

När du använder en LLM betalar du för antal tokens du spenderar
- Prompt (input): den text du skickar in
- Completion (output): det svar modellen genererar
- Input och output kan ha olika priser

**Exempel**

- GPT-3.5-turbo (en något äldre modell) kostar ca 10 kr för 1 miljon tokens om de är jämt fördelade över input och output
- GPT-4.1 kostar ca 200 kr för 1 miljon tokens om de är jämt fördelade över input och output

**API'er är stateless**

API'ier är 'stateless' vilket innebär att de kommer inte ihåg något mellan anropen. Det innebär att i en konversation med en modell
skickas hela konversationen fram och tillbaka vid varje anrop. En längre konversation kan bränna många tokens.

**Komma igång med API'er**
- skapa ett konto hos någon komersiell aktör
- ange betalningsinformation
- generera en API-nyckel
- ladda ner klientbibliotek för vald tjänst
- skriv kod, ange API-nyckel och modell i anropet
